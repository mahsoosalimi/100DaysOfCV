#**Image Processing** with different thresholds

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib auto
from google.colab import drive
```

```python
# read an image
img = cv2.imread('./Blueberry1.jpeg', 0)
```

```python
thr, mask = cv2.threshold(img, 200, 1, cv2.THRESH_BINARY)
print('Threshold used:', thr)
```

```python
adapt_mask = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                   cv2.THRESH_BINARY_INV, 11, 10)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1, 3, 1)
plt.axis('off')
plt.title('original')
plt.imshow(img, cmap='gray')
plt.subplot(1, 3, 2)

plt.axis('off')
plt.title('binary threshold')
plt.imshow(mask, cmap='gray')
plt.subplot(1, 3, 3)

plt.axis('off')
plt.title('adaptive threshold')
plt.imshow(adapt_mask, cmap='gray')
plt.tight_layout()
plt.show()
```
