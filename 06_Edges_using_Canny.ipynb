#**Finding edges using Canny algorithm**



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
from google.colab import drive
```

```python
# read an image
img = cv2.imread('./Blueberry1.jpeg')

img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
```

```python
edges = cv2.Canny(img, 200, 100)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.axis('off')
plt.imshow(img[:,:,[2,1,0]])

plt.subplot(1, 2, 2)
plt.axis('off')
plt.imshow(edges, cmap='gray')
plt.tight_layout()
plt.show()
```
